## Where is Euclidean distance used?


Euclidean distance calculates the distance between two real-valued vectors. You are most likely to use Euclidean distance when calculating the distance between two rows of data that have numerical values, such a floating point or integer values.



```python
# importing the library
from scipy.spatial import distance

# defining the points
point_1 = (1, 2, 3)
point_2 = (4, 5, 6)
point_1, point_2
```

```python
# computing the euclidean distance

euclidean_distance = distance.euclidean(point_1, point_2)

print('Euclidean Distance b/w', point_1, 'and', point_2, 'is: ', euclidean_distance)
```

```python
# Python code to find Euclidean distance using square() and sum() methods
# Import NumPy Library
import numpy as np

# initializing points in numpy arrays
point1 = np.array((1, 2, 3))
point2 = np.array((4, 5, 6))

# finding sum of squares
sum_vectors = np.sum(np.square(point1 - point2))
 
# performing the squareroot and printing Euclidean distance
print(np.sqrt(sum_vectors))
```

# 2. Manhattan Distance


plt.show()

### We use Manhattan distance, also known as city block distance, or taxicab geometry if we need to calculate the distance between two data points in a grid-like path.

```python
# computing the manhattan distance

manhattan_distance = distance.cityblock(point_1, point_2)

print('Manhattan Distance b/w', point_1, 'and', point_2, 'is: ', manhattan_distance)
```

```python
from math import sqrt

#create function to calculate Manhattan distance 
def manhattan(p, q):
    return sum(abs(val1-val2) for val1, val2 in zip(p,q))
 
#define vectors
p = [1, 2, 3]
q = [4, 5, 6]

#calculate Manhattan distance between vectors
manhattan(p, q)
```

# 3. Minkowski Distance

Minkowski distance calculates the distance between two real-valued vectors. It is a generalization of the Euclidean and Manhattan distance measures and adds a parameter, called the order or p, that allows different distance measures to be calculated.

The p parameter of the Minkowski Distance metric of SciPy represents the order of the norm. When the order(p) is 1, it will represent Manhattan Distance and when the order in the above formula is 2, it will represent Euclidean Distance.

plt.show()



```python
# computing the minkowski distance

minkowski_distance = distance.minkowski(point_1, point_2, p=3)

print('Minkowski Distance b/w', point_1, 'and', point_2, 'is: ', minkowski_distance)
```

# 4. Hamming Distance

larger the Hamming Distance between two strings, more dissimilar will be those strings (and vice versa).

```python
# defining two strings

string_1 = 'euchidean'
string_2 = 'manhattan'
```

plt.show()



```python
# computing the hamming distance

hamming_distance = distance.hamming(list(string_1), list(string_2))*len(string_1)

print('Hamming Distance b/w', string_1, 'and', string_2, 'is: ', hamming_distance)
```
